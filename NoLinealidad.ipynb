```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
```

```python
T = np.linspace(0,400)
V = 296456 + 1497.6*T + 0.31787*(T**2) - 0.0031847*(T**3) + 1.5721e-6*(T**4)
```

# Voltaje real 

```python
plt.plot(T, V, '-r', label='V = 296456 + 1497.6*T + 0.31787*(T**2) - 0.0031847*(T**3) + 1.5721E-6*(T**4)')
plt.title('Graph of V')
plt.xlabel('T', color='#1C2833')
plt.ylabel('V(uv)', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
T
```

# Voltaje Ideal

```python
x = Symbol('x')
Vder = 296456 + 1497.6*x + 0.31787*(x**2) - 0.0031847*(x**3) + 1.5721e-6*(x**4)
Videal = Vder.diff(x)
```

```python
fun1 = Vder.subs({x: T[1]})
fun2 = Videal.subs({x: T[1]})
Vrec = fun1 + (x-T[1])* fun2
```

```python
valurec = np.zeros(T.size)
for i in range(T.size):
    valurec[i] = Vrec.subs({x: T[i]})
```

```python
valurec
```

```python
fig, ax = plt.subplots()
ax.plot(T, V, '-r', label=Vder)
ax.plot(T, valurec, '-b', label=Vrec)


ax.legend(loc='upper left')
```
